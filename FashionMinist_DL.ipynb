```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
x_train = train_df.iloc[:,1:].to_numpy()   #select columns except (label)-> numpy array
x_train = x_train.reshape([-1,28,28,1])     #reshape for 28*28 1-> greyscale
x_train = x_train / 255                     #divide to scale the values
```

```python
y_train = train_df.iloc[:,0].to_numpy()   #select 1st clo->label ->numpy array
```

```python
x_test = test_df.iloc[:,1:].to_numpy()
x_test = x_test.reshape([-1,28,28,1])
x_test = x_test / 255
```

```python
y_test = test_df.iloc[:,0].to_numpy()
```

```python
#visulaize sample images -> class labels
plt.figure(figsize=(10,10))
for i in range(25):        #loop over 1st 25 training img
    plt.subplot(5,5,i+1)    #create 5*5 grid of subplot
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)    #render the black and white imgs 
    plt.xlabel(class_names[y_train[i]])           #display class names 
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D,Dropout
```

```python
model = Sequential()


model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(10, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=1200,validation_split=0.05)
```

```python
evaluation = model.evaluate(x_test,y_test)
```

```python
print(f"Accuracy: {evaluation[1]}")
```

```python
y_probas = model.predict(x_test)
```

```python
y_pred = y_probas.argmax(axis=-1)
```

```python
y_pred
```

```python
plt.figure(figsize=(10,10),)
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[i], cmap=plt.cm.binary)
#     plt.xlabel(f"True Class:{y_test[i]}")
    plt.title(f"Pred:{class_names[y_pred[i]]}")
plt.show()
```

```python
cr = classification_report(y_test, y_pred, target_names=class_names)
print(cr)
```
